```python
Rest_Data = pd.read_csv("Rest-Data.csv", encoding='latin1')
# Defination of cuisine_counter function
def cuisine_counter(cuisines):
    # Spliting the cuisines string by commas 
    if pd.notnull(cuisines):
        #returning cuisine count
        return len(cuisines.rsplit(',')) 


# Createing  new column named "CuisineCount" and applying the cuisine_counter function
Rest_Data['CuisineCount'] = Rest_Data['Cuisines'].apply(cuisine_counter)
#Display first five instance
Rest_Data.head()
```

```python
#ANOVA test to determine if there are significant differences in ratings based on whether a restaurant is delivering now or not 
#ANOVA test is to find out relation between catogorical and continous variable statistically

from scipy import stats
continuous_variable = Rest_Data['Rating']
categorical_variable = Rest_Data['Is delivering now']

# Create a dictionary to store data 
data_by_group = {}
for delivery in categorical_variable.unique():
    data_by_group[delivery] = continuous_variable[categorical_variable == delivery]

# Perform the ANOVA test
f_statistic, p_value = stats.f_oneway(*data_by_group.values())

# Print the results
print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Check if the p-value is less than the significance level (we have considered 0.05)
if p_value < 0.05:
    print("There are significant differences in ratings based on if delivering now .")
else:
    print("There are no significant differences in ratings based on if delivering now .")
```

```python
#Treating Outliers
#Replacing outliers in data frame column with mean value using z_score threshold
def replace_outliers_with_mean(df,column_name, z_score_threshold=4):
 # Calculate the mean and standard deviation of the column
    mean_value = df[column_name].mean()
    std_dev = df[column_name].std()
 # Define the lower and upper bounds for outliers
    lower_bound = mean_value - z_score_threshold * std_dev
    upper_bound = mean_value + z_score_threshold * std_dev
# Replace outliers with the mean value using .loc
    df.loc[(df[column_name] < lower_bound) | (df[column_name] > upper_bound), column_name] = mean_value

# Replace outliers in the 'Votes' column of the 'New_Rest_Data' DataFrame with a Z-score threshold of 6
replace_outliers_with_mean(New_Rest_Data, 'Votes', z_score_threshold=6)
New_Rest_Data

```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Initialize the Decision Tree Regressor
tree_depth = range(1, 10)
tree_test_accuracy = []
tree_train_accuracy = []

for tb in tree_depth: 
    # Initialize the Decision Tree Regressor with a specified maximum depth
    tree_regressor = DecisionTreeRegressor(max_depth=tb, random_state=523)

    # Train the model on the training data
    tree_regressor.fit(X_train, y_train)



    # Make predictions on the training and testing data
    y_train_pred = tree_regressor.predict(X_train)
    y_test_pred = tree_regressor.predict(X_test)

    # Evaluate the model's performance using regression metrics
    mae_train = mean_absolute_error(y_train, y_train_pred)
    mae_test = mean_absolute_error(y_test, y_test_pred)
    mse_train = mean_squared_error(y_train, y_train_pred)
    mse_test = mean_squared_error(y_test, y_test_pred)
    rmse_train = mean_squared_error(y_train, y_train_pred, squared=False)
    rmse_test = mean_squared_error(y_test, y_test_pred, squared=False)
    r2_train = r2_score(y_train, y_train_pred)
    r2_test = r2_score(y_test, y_test_pred)
    
    # Append the R-squared (R^2) scores to the accuracy lists
    tree_test_accuracy.append(r2_test)
    tree_train_accuracy.append(r2_train)  
    
    # Print evaluation metrics if depth equals 7
    if tb == 7:
        print("Training Set Metrics:")
        print("Mean Absolute Error (MAE):", mae_train)
        print("Mean Squared Error (MSE):", mse_train)
        print("Root Mean Squared Error (RMSE):", rmse_train)
        print("R-squared (R^2) Score:", r2_train)

        print("\nTesting Set Metrics:")
        print("Mean Absolute Error (MAE):", mae_test)
        print("Mean Squared Error (MSE):", mse_test)
        print("Root Mean Squared Error (RMSE):", rmse_test)
        print("R-squared (R^2) Score:", r2_test)
        
    else:
        continue

# Plot the training and testing accuracy as a function of tree depth
plt.plot(tree_depth, tree_train_accuracy, label="Training Accuracy")
plt.plot(tree_depth, tree_test_accuracy, label="Testing Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Depth of Tree")
plt.legend()

```
