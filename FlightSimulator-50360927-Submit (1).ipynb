```python
import random
file = open("FlightSimulator.txt", "w")

# List of 5 planes initialized to 100 gallons fuel level
planes = {1:100,2:100,3:100,4:100,5:100}


#function to display number of planes and fuel levels
def display_planes():
    num_planes = len(planes)

    #Assigning print statements to a variable called message 
    message = "------------------------------------\nThere are " + str(num_planes) + " planes in the Sky.\n"
   #Writing message to file called FlightSimulator.txt
    file.write(message)
   #Write the same message to console 
    print(message)
    
# Loop to display all planes with numbers and fuel levels
    for key in planes:
        plane_number = key
        fuel_level = planes[plane_number]
        
        message = "Plane " + str(plane_number) + " - Fuel Level: " + str(fuel_level) + " Gallons\n"
        file.write(message)
        print(message)
        


#function to land planes and display which plane landed based on user input 
def land_plane():
    if len(planes) > 0:
        plane_to_land = int(input("Enter the Plane Number to Land: "))
        print(planes.keys())
        
        message = "Enter the Plane Number to Land: " + str(plane_to_land) + "\n"
        file.write(message)
        
        
        #checking condition to check wheather plane is in the sky or already landed
        if plane_to_land in planes.keys():
            planes.pop(plane_to_land)
            
            message = "Plane " + str(plane_to_land) + " has landed\n"
            file.write(message)
            print(message)
        else:
           
            message = "Invalid plane number.\n"
            file.write(message)
            print(message)

    else:
        
        message = "No planes left to land.\n"
        file.write(message)
        print(message)


#function to reduce fuel from planes
def reduce_fuel():
    if len(planes) > 0:
        #taking user to give count on how many planes should reduce fuel level 
        count = int(input("How many Planes Should have their Fuel Level Reduced: "))
        
        message = "How many Planes Should have their Fuel Level Reduced: " + str(count) + "\n"
        file.write(message)
        
        if 1 <= count <= len(planes):
            for i in range(count): 
                #calling random to choose random integer based on user input
                plane_index = random.choice(list(planes.keys()))
                #checking fuel level and reducing it by 15 gallons
                fuel_level = planes[plane_index]
                if fuel_level <= 15:
                    fuel_reduction = fuel_level
                else:
                    fuel_reduction = 15
                planes[plane_index] = planes[plane_index]-fuel_reduction
                
                message = "Plane " + str(plane_index) + " Fuel Level reduced by " + str(fuel_reduction) + " Gallons\n"
                file.write(message)
                print(message)
        else:
            
            message = "Invalid number of planes.\n"
            file.write(message)
            print(message)
    else:
        
        message = "No planes left to reduce fuel.\n"
        file.write(message)
        print(message)
        
        


#Definition of main
def main():
    while True:
        #calling display planes function
        display_planes() 
        
        #Displaying options for user
        print("a. Land the Plane")
        print("b. Reduce Fuel")
        print("c. Quit Program")
        
        #Taking users choice 
        choice = input("Enter your Choice: ")
        message = "Enter your Choice: " + choice + "\n"
        file.write(message)
        
        #Checking the choice and calling respective function to perform corresponding actions 
        if choice == "a":
            land_plane()

        elif choice == "b":
            reduce_fuel()

        elif choice == "c":
            
            message = "Exiting the program.\n"
            file.write(message)
            print(message)
            file.close()
            break

        else:
            
            message = "Invalid choice. Enter a valid option.\n"
            file.write(message)
            print(message)

# Calling the main function           
main()  
```
